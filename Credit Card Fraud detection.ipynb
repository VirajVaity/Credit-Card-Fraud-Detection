```python
import pandas as pd
```

```python
data=pd.read_csv("C:/Users/viraj/Downloads/creditcardfraud.csv")
```

```python
data.drop(['Time'],inplace=True,axis=1) #dropping Time coloum
```

```python
data
```

```python
from sklearn.preprocessing import StandardScaler
import numpy as np
```

```python
Scalar=StandardScaler() #using standard scalar for scaling data
```

```python
Scaled_data=Scalar.fit_transform(data[['Amount']])
```

```python
Scaled_data
```

```python

from sklearn.model_selection import train_test_split
```

```python
data.columns
```

```python
X=data.drop(['Class','Amount'],axis=1) #seperating features 
```

```python
y=data['Class']#target variable
```

```python

xtrain,xtest,ytrain,ytest=train_test_split(X,y,test_size=.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression 
```

```python
LR=LogisticRegression(solver = "newton-cg") 
```

```python
LR.fit(xtrain,ytrain)
```

```python
ypred = LR.predict(xtest)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(ypred,ytest)
```

```python
#Confusion MAtrix
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
cm=confusion_matrix(ypred,ytest)
```

```python
cr=classification_report(ypred,ytest)
```

```python
cm
```

```python
cr
```

```python
#SVM
```

```python
from sklearn.svm import SVC
```

```python
svc=SVC()
```

```python
svc.fit(xtrain,ytrain)
```

```python
ypred=svc.predict(xtest)
```

```python
ypred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(ypred,ytest)
```

```python
# Implementing Grid Search 
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid_logreg = {
    'C':[0.1, 1, 10, 100],
    'penalty': ['l1', 'l2']
}
```

```python
grid_search_logreg = GridSearchCV(LR, param_grid_logreg, cv=5, scoring='accuracy')
grid_search_logreg.fit(xtrain, ytrain)
```

```python
best_params_logreg = grid_search_logreg.best_params_
```

```python
y_pred_logreg = grid_search_logreg.predict(xtest)
accuracy_logreg = accuracy_score(ytest, y_pred_logreg)
```

```python
print("Best Hyperparameters for Logistic Regression: ", best_params_logreg)
print("Accuracy on Test Set: ", accuracy_logreg)
```

```python
#Grid search on SVC
```

```python
param_grid_svm = {
    'C': [0.1, 1, 10, 100],
    'kernel': ['linear', 'rbf', 'poly'],
    'gamma': ['scale', 'auto']
}
```

```python
grid_search_svm = GridSearchCV(svc, param_grid_svm, cv=5, scoring='accuracy')
grid_search_svm.fit(xtrain, ytrain)
```

```python
best_params_svm = grid_search_svm.best_params_
```

```python
y_pred_svm = grid_search_svm.predict(xtest)
accuracy_svm = accuracy_score(ytest, y_pred_svm)

print("Best Hyperparameters for SVM: ", best_params_svm)
print("Accuracy on Test Set: ", accuracy_svm)
```

```python
# Training Logistic Regression with Optimal Hyperparameters
optimal_logreg = LogisticRegression(C=best_params_logreg['C'], penalty=best_params_logreg['penalty'])
optimal_logreg.fit(xtrain, ytrain)

# Evaluate Logistic Regression on the test set
y_pred_logreg = optimal_logreg.predict(xtest)
accuracy_logreg = accuracy_score(ytest, y_pred_logreg)

print("Logistic Regression - Accuracy on Test Set: ", accuracy_logreg)

# Training SVM with Optimal Hyperparameters
optimal_svm = SVC(C=best_params_svm['C'], kernel=best_params_svm['kernel'], gamma=best_params_svm['gamma'])
optimal_svm.fit(xtrain, ytrain)

# Evaluate SVM on the test set
y_pred_svm = optimal_svm.predict(xtest)
accuracy_svm = accuracy_score(ytest, y_pred_svm)

print("SVM - Accuracy on Test Set: ", accuracy_svm)
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score

# Assuming you already have y_test, y_pred_logreg, and y_pred_svm from the previous code snippets

# Calculate additional metrics for comparison
precision_logreg = precision_score(ytest, y_pred_logreg, average='weighted')
recall_logreg = recall_score(ytest, y_pred_logreg, average='weighted')
f1_logreg = f1_score(ytest, y_pred_logreg, average='weighted')

precision_svm = precision_score(ytest, y_pred_svm, average='weighted')
recall_svm = recall_score(ytest, y_pred_svm, average='weighted')
f1_svm = f1_score(ytest, y_pred_svm, average='weighted')

# Display the results
print("Logistic Regression Metrics:")
print("Accuracy: {:.4f}".format(accuracy_logreg))
print("Precision: {:.4f}".format(precision_logreg))
print("Recall: {:.4f}".format(recall_logreg))
print("F1 Score: {:.4f}".format(f1_logreg))
print("\nSVM Metrics:")
print("Accuracy: {:.4f}".format(accuracy_svm))
print("Precision: {:.4f}".format(precision_svm))
print("Recall: {:.4f}".format(recall_svm))
print("F1 Score: {:.4f}".format(f1_svm))
```
